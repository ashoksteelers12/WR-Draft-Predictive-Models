```python
# J.J. Arcega-Whiteside
filename = 'Game-Stats-2019/' + WR_Prospects_2019[0][0] + '.csv'
df = pd.read_csv(filename)
new_df = df[['Rec','Yds','TD']].values

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(new_df[:, 0], new_df[:, 1], new_df[:,2])
plt.show()

ph = np.zeros((len(new_df), 3))
for i in range(len(new_df)):
    ph[i][0] = new_df[i][0]
    ph[i][1] = new_df[i][1]
    ph[i][2] = new_df[i][2]
diagrams = ripser(ph, thresh=999, maxdim=2)['dgms']
plot_diagrams(diagrams, xy_range=[-2,5,-1,5], show=True)
```
