```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from ripser import ripser
from persim import plot_diagrams
```

## 2018 WR Prospects

```python
%store -r WR_Prospects_2018
```

```python
df_2018 = pd.DataFrame(WR_Prospects_2018, columns = ['Name','Conference','College','Year','Grade','Round','Class'])
```

```python
gcr_2018 = df_2018[['Grade', 'Class', 'Round']].values
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(gcr_2018[:, 0], gcr_2018[:, 1], gcr_2018[:,2])
plt.show()
```

```python
ph_2018 = np.zeros((len(WR_Prospects_2018), 3))
for i in range(len(WR_Prospects_2018)):
    ph_2018[i][0] = WR_Prospects_2018[i][4]
    ph_2018[i][1] = WR_Prospects_2018[i][5]
    ph_2018[i][2] = WR_Prospects_2018[i][6]
diagrams = ripser(ph_2018, thresh=999, maxdim=2)['dgms']
plot_diagrams(diagrams, xy_range=[-2,5,-1,5], show=True)
```
